# Application Task

## Age Estimation
The following task consist of designing, training and evaluating a neural network for age estimation of face images. For this task there is not any predefined code and you are completely free to prototype it by yourself. Instead of solving the age prediction task as a label classification problem, we want to estimate the age from a predicted distribution (see the figure below, where the age label $43$ is plotted as a distribution $\mathcal{N}(43,9)$).

![alt text](https://drive.google.com/uc?id=125fuZLgtaszXfV9FNppTBDUrGR2EJ9O5)

Since we are dealing with different distributions the cost to minimize could be the Kullbackâ€“Leibler divergence between the ground truth age distribution and the predicted one

$$D_{KL}\left(y \hspace{0.1cm} \| \hspace{0.1cm} p \right) = \displaystyle\sum_{j}y_{j}\log\left(\frac{y_{j}}{p_{j}}\right),$$

where $y_j$ is the $j^{th}$ value in the ground truth label distribution and $p_j$ is the $j^{th}$ of the predicted distribution. This leads to a loss function defined as,

$$\mathcal{L}\left(y, p \right) = -\displaystyle\sum_{j}y_{j}\log\left(p_{j}\right).$$

To evaluate the age estimation we want to use the error defined as,

$$\epsilon(n, \mu, \sigma) = 1 - \exp\left(-\frac{\left(n-\mu\right)^{2}}{2\sigma^{2}} \right),$$

where $n$ is the predicted age, and $\mu$ and $\sigma^{2}$ the target's mean and variance, respectively.

For training the network we provide a sample of the UTK Face dataset [1], which already has age labels (written in the first integer in each image file name) for aligned and cropped face images.

Regarding possible network architectures, please use a simple predefined backbone network like LeNet for the resolution of the task.

We are not interested in you making the most accurate model. As far as you justify what's happening we'll be satisfied.
At the evaluation moment we will focus in your overall workflow, i.e. in how you approach such an open task. Last but not least, we are also interested in the overall coding skills, so please don't drop the ball in that aspect.

Here are some example questions for you to guide your resolution:
- Briefly describe how the $\epsilon$ error behaves in the limit cases
- What can you say about the data sample?
- Can you justify the behaviour of your training $\epsilon$ error. What's its first value? Why?
- Is your model learning?
- Further improvements?
